### 2. Remove HTML Tags: Since this data has been scraped from the web, it contains a number of HTML tags that indicate components of the DOM but aren't useful for analysis. Remove any HTML tags, but keep the contents of the comment. For example, "< div > Hi, Mike! < /div >" will become Hi, Mike!
### (Aaron advised we solved by starting with 2)

```python
#Number 2
#Import re function(https://docs.python.org/3/library/re.html)
import re
#Remove as many special and standard HTML style characters(https://lzone.de/examples/Python%20re.sub)
def HTMLRevision(df):
    #df['comment_msg'] = df['comment_msg'].apply(lambda x: str(x).lstrip())
    df['comment_msg'] = df['comment_msg'].apply(lambda x: str(x).strip())
    df['comment_msg'] = df['comment_msg'].apply(lambda x: re.sub(r'\r*',"", str(x)))
    df['comment_msg'] = df['comment_msg'].apply(lambda x: re.sub(r'<.*?>',"",str(x)))
#Return and check re edit changes
    return df    
clean_html_df = HTMLRevision(df)
clean_html_df.head(10)
```

### 1. Remove spam: Some comments on social media contain spam and aren't useful for our analysis. We can determine comments that are spam by removing any comments containing a web address or any of the following keywords: App, FREE, %20, Check out my page. Note that these keywords are NOT case sensitive (e.g., you should remove comments containing APP, app, App, APp, etc.).

```python
#Number 1
#Remove web address content(http.www.com)/App, FREE, %20, Check out my page.(Not case sensitive, remove all variations)
def SpamRevision(clean_html_df):
    clean_html_df['comment_msg'] = clean_html_df['comment_msg'].apply(lambda x: x.lower())
    spam_comments_remove = ['app', 'free', '%20', 'check out my page', 'http://', 'www.', '.com']
    clean_html_df = clean_html_df[clean_html_df.comment_msg.str.contains('|'.join(spam_comments_remove)) == False]
#Return and check re edit changes    
    return clean_html_df
clean_html_spam_df = SpamRevision(df)
clean_html_spam_df.head(10)
```

### 3. Remove Missing Data: The data collection script used to scrape MySpace inserted null values in cases where the data collection was unable to process meaningful content. Remove all rows where the message is null.

```python
#Number 3
#Remove null values given to unprocessable content.(null/nan)
def NullRevision(clean_html_spam_df):
    clean_html_spam_df = clean_html_spam_df[df['comment_msg'] != "nan"]
    clean_html_spam_df = clean_html_spam_df[df['comment_msg'] != "null"]
    clean_html_spam_df = clean_html_spam_df[df['comment_msg'] != ""]
    df.dropna()
#Return and check null/nan edit changes    
    return clean_html_spam_df
final_clean_df = NullRevision(clean_html_spam_df)
final_clean_df.head(10)
```
